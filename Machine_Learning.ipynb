##Q1 Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).

<br>

Ans :-
1. **Artificial Intelligence (AI)**  
  AI is the broad field of making machines perform tasks that normally require human intelligence.  
  It covers problem-solving, understanding language, decision-making, and perception.  
   **Example:** A chatbot that answers customer questions.

2. **Machine Learning (ML)**  
   ML is a part of AI that focuses on teaching machines to learn from data instead of being programmed with exact rules.  
   **Example:** A spam filter that learns to detect unwanted emails.

3. **Deep Learning (DL)**  
   DL is a type of ML that uses artificial neural networks with many layers to process and learn from large, complex datasets.  
   **Example:** A system that can recognize objects in images.

4. **Data Science (DS)**  
   DS is about extracting insights from data using statistics, programming, and domain expertise.  
   It often uses AI and ML techniques as tools.  
   **Example:** Analyzing sales data to predict next month’s demand.

**Summary:**  
- **AI** is the biggest field, covering everything about making machines intelligent.  
- **ML** is a subset of AI that learns from data.  
- **DL** is a specialized form of ML for very complex problems.  
- **DS** overlaps with AI/ML but focuses on data analysis and insight generation.

##Q2 What are the types of machine learning? Describe each with one real-world example.

<br>

Ans :- Machine learning is generally divided into three main types:

1. **Supervised Learning**  
   - The model learns from labeled data (input-output pairs).  
   - Goal: Predict the output for new inputs.  
   - **Example:** Predicting house prices based on features like size, location, and number of rooms.

2. **Unsupervised Learning**  
   - The model works with unlabeled data and finds hidden patterns or groupings.  
   - Goal: Discover structure in data without predefined outputs.  
   - **Example:** Customer segmentation in marketing, grouping customers with similar buying habits.

3. **Reinforcement Learning**  
   - The model learns by interacting with an environment and receiving rewards or penalties for its actions.  
   - Goal: Maximize cumulative rewards through trial and error.  
   - **Example:** A self-driving car learning to navigate roads by trial and error, improving with each drive.

##Q3. Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.**

<br>

Ans :-
1. **Overfitting**  
   - Happens when a model learns the training data too well, including noise and random fluctuations.  
   - The model performs very well on training data but poorly on new, unseen data.  
   - **Example:** A decision tree that memorizes every detail of the training set and fails to generalize.

2. **Underfitting**  
   - Happens when a model is too simple to capture the underlying patterns in the data.  
   - The model performs poorly on both training and test data.  
   - **Example:** Using a straight line to fit data that has a complex curve.

3. **Bias-Variance Tradeoff**  
   - **Bias**: Error from oversimplifying the model (can cause underfitting).  
   - **Variance**: Error from making the model too complex (can cause overfitting).  
   - Tradeoff: Increasing model complexity reduces bias but increases variance, and vice versa.  
   - Goal: Find a balance where both bias and variance are low for the best generalization.

##Q4 What are outliers in a dataset, and list three common techniques for handling them

Ans :-

**Outliers** are data points that differ greatly from the majority of the data.  
They may occur due to measurement errors, data entry mistakes, or genuine unusual events.  
Outliers can distort statistical results and reduce model accuracy.

**Three common techniques for handling outliers:**

1. **Removal**  
   - Delete the outlier values if they are due to errors or irrelevant to analysis.  
   - Example: Removing a height value of 300 cm due to a data entry mistake.

2. **Transformation**  
   - Apply mathematical transformations (like log or square root) to reduce the effect of extreme values.

3. **Capping (Winsorizing)**  
   - Replace extreme values with a set maximum or minimum threshold.  
   - Example: Replace all values above the 95th percentile with the 95th percentile value.

##Q5 Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.

Ans :-

**Handling missing values** is important to maintain data quality and ensure accurate analysis.  
The process typically involves:

1. **Identifying missing values**  
   - Use functions like `isnull()` or `isna()` in Python (Pandas) to detect them.

2. **Deciding on a strategy**  
   - Remove rows/columns with too many missing values.
   - Fill (impute) missing values using suitable techniques.

3. **Applying imputation methods**  
   - Choose techniques based on the data type and distribution.

**One imputation technique for numerical data:**  
- **Mean imputation**: Replace missing numbers with the average value of that column.

**One imputation technique for categorical data:**  
- **Mode imputation**: Replace missing values with the most frequent category in that column.

###Q6 Write a Python program that:

● Creates a synthetic imbalanced dataset with make_classification() from
sklearn.datasets.

● Prints the class distribution.

```python
from sklearn.datasets import make_classification
import numpy as np

# Create an imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=5, n_informative=3,
                           n_redundant=0, n_clusters_per_class=1,
                           weights=[0.9, 0.1], random_state=42)

# Print class distribution
unique, counts = np.unique(y, return_counts=True)
class_distribution = {int(k): int(v) for k, v in zip(unique, counts)}

print("Class Distribution:", class_distribution)
```

###Q8 Write a Python script to:

● Generate 1000 samples from a normal distribution.

● Introduce 50 random missing values.

● Fill missing values with the column mean.

● Plot a histogram before and after imputation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(1000, 50, replace=False)
data[missing_indices] = np.nan

# Step 3: Create DataFrame
df = pd.DataFrame({'Values': data})

# Step 4: Plot histogram before imputation
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(df['Values'], bins=20, color='skyblue', edgecolor='black')
plt.title("Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Step 5: Fill missing values with column mean
mean_value = df['Values'].mean()
df['Values'] = df['Values'].fillna(mean_value)

# Step 6: Plot histogram after imputation
plt.subplot(1, 2, 2)
plt.hist(df['Values'], bins=20, color='lightgreen', edgecolor='black')
plt.title("After Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

 ### Q10 You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains:

● Missing ages,

● Outliers in transaction amount,

● A highly imbalanced target (fraud vs. non-fraud),

● Categorical variables like payment method.

Explain the step-by-step data preparation plan you’d follow before training a machine learning model. Include how you’d address missing data, outliers, imbalance, and encoding.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from imblearn.over_sampling import SMOTE

# Step 1: Simulate dataset
np.random.seed(42)
data = pd.DataFrame({
    'age': np.random.choice([25, 30, 35, 40, np.nan], size=100),
    'transaction_amount': np.random.normal(500, 200, 100),
    'payment_method': np.random.choice(['Card', 'UPI', 'Wallet', 'PayPal'], size=100),
    'is_fraud': np.random.choice([0, 1], size=100, p=[0.9, 0.1])
})

# Introduce outliers
data.loc[np.random.choice(data.index, 5, replace=False), 'transaction_amount'] = 5000

print("=== Original Data Sample ===\n", data.head(), "\n")

# Step 2: Missing values visualization
plt.figure(figsize=(8,3))
plt.hist(data['age'].dropna(), bins=5, color='skyblue', edgecolor='black')
plt.title("Age Distribution (Before Imputation)")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Step 3: Handle missing ages (mean imputation)
age_imputer = SimpleImputer(strategy='mean')
data['age'] = age_imputer.fit_transform(data[['age']])

# Step 4: Outlier visualization before capping
plt.figure(figsize=(8,3))
plt.hist(data['transaction_amount'], bins=20, color='salmon', edgecolor='black')
plt.title("Transaction Amount (Before Capping)")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()

# Step 5: Cap outliers at 99th percentile
cap_value = data['transaction_amount'].quantile(0.99)
data['transaction_amount'] = np.where(
    data['transaction_amount'] > cap_value, cap_value, data['transaction_amount']
)

# Step 6: Outlier visualization after capping
plt.figure(figsize=(8,3))
plt.hist(data['transaction_amount'], bins=20, color='lightgreen', edgecolor='black')
plt.title("Transaction Amount (After Capping)")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()

# Step 7: One-hot encode categorical variable
encoder = OneHotEncoder(drop='first')
encoded_payment = encoder.fit_transform(data[['payment_method']])
encoded_df = pd.DataFrame(encoded_payment.toarray(), columns=encoder.get_feature_names_out(['payment_method']))
data = pd.concat([data.drop(columns=['payment_method']), encoded_df], axis=1)

# Step 8: Class imbalance visualization before SMOTE
plt.figure(figsize=(5,3))
data['is_fraud'].value_counts().plot(kind='bar', color=['skyblue','salmon'], edgecolor='black')
plt.title("Class Distribution (Before SMOTE)")
plt.xlabel("Class (0=Non-Fraud, 1=Fraud)")
plt.ylabel("Count")
plt.show()

# Step 9: Handle class imbalance with SMOTE
X = data.drop(columns=['is_fraud'])
y = data['is_fraud']
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Step 10: Class imbalance visualization after SMOTE
plt.figure(figsize=(5,3))
pd.Series(y_resampled).value_counts().plot(kind='bar', color=['skyblue','salmon'], edgecolor='black')
plt.title("Class Distribution (After SMOTE)")
plt.xlabel("Class (0=Non-Fraud, 1=Fraud)")
plt.ylabel("Count")
plt.show()

# Step 11: Final processed sample
processed_df = pd.DataFrame(X_resampled, columns=X.columns)
print("=== Processed Data Sample ===\n", processed_df.head())
```
